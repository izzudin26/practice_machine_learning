```python
import pandas as pd
```

```python
df = pd.read_csv('diabetes_prediction_dataset.csv')
df.info()
```

```python
df.isna().sum()
```

```python
df.head()
```

```python
df.gender.unique()
```

```python
df.gender.replace(['Female', 'Male', 'Other'], [0, 1, 2], inplace=True)
df.head()
```

```python
df.smoking_history.unique()
df.smoking_history.replace(['never', 'No Info', 'current', 'former', 'ever', 'not current'], [0, 1, 2, 3, 4, 5], inplace=True)
df.head()
```

```python
df.info()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# ploting
f, ax = plt.subplots(figsize=(8, 8))
sns.heatmap(df.corr(), annot=True, ax=ax)
plt.show()
```

```python
df.describe()
```

```python
df['diabetes'] = df['diabetes'].astype('int64')
df['heart_disease'] = df['heart_disease'].astype('int64')
df['age'] = df['age'].astype('int64')
df.head()
```

```python
from sklearn.model_selection import train_test_split
# spliting

inp = df[['age', 'hypertension', 'heart_disease', 'smoking_history', 'bmi', 'HbA1c_level', 'blood_glucose_level']]
out = df[['diabetes']]

x_train, x_test, y_train, y_test = train_test_split(inp, out, test_size=0.2, random_state=20)
```

```python
# Training
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(solver='lbfgs', max_iter=200)
model.fit(x_train, y_train)
```

```python
accurate = model.score(x_test, y_test)
print(f"Accurate model {accurate}")
```

```python
sample = [80, 0, 1, 0, 25.19, 6.6, 140]
print(model.predict([sample]))
```
